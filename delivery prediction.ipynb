```python
df.dropna(subset=['actual_delivery_time', 'created_at'], inplace=True)
```

```python
df['market_id'].fillna(-1,inplace=True)
df['store_primary_category'].fillna('Unknown', inplace=True)
df['order_protocol'].fillna(-1, inplace=True)
```

```python
df['total_onshift_dashers'].fillna(df['total_onshift_dashers'].median(), inplace=True)
df['total_busy_dashers'].fillna(df['total_busy_dashers'].median(), inplace=True)
df['total_outstanding_orders'].fillna(df['total_outstanding_orders'].median(), inplace=True)
df['estimated_store_to_consumer_driving_duration'].fillna(df['estimated_store_to_consumer_driving_duration'].median(), inplace=True)
```

```python
df['hour'] = df['created_at'].dt.hour
df['day_of_week'] = df['created_at'].dt.dayofweek
df['month'] = df['created_at'].dt.month
df = pd.get_dummies(df, columns=['hour', 'day_of_week', 'month', 'store_primary_category'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=40,random_state=40)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```
